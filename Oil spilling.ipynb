```python
import xgboost as xgb
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
import pandas as pd
from sklearn.model_selection import train_test_split

# Load the dataset
data = pd.read_csv('vessel_data_demo.csv')

# Fill missing values
data = data.ffill()

# Check class distribution
print("Class distribution:\n", data['Navigational Status'].value_counts())

# Define features and target
X = data[['Speed (kn)', 'Course (°)', 'Draught (m)']]  # Feature columns
y = data['Navigational Status']  # Target column

# Encode target if it's categorical
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Split the data into train and test sets (stratified)
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# Initialize the XGBoost Classifier
model = xgb.XGBClassifier()

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Print the confusion matrix with specified labels
conf_matrix = confusion_matrix(y_test, y_pred, labels=le.transform(le.classes_))
print("Confusion Matrix:")
print(conf_matrix)

# Print the classification report with specified labels
class_report = classification_report(y_test, y_pred, target_names=le.classes_)
print("Classification Report:")
print(class_report)
```
